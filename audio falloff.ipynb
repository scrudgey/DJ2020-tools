```python
def plot(function):
    fig = plt.Figure(figsize=(5, 5))
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    y = [function(x) for x in x_range]
    ax.plot(x_range, y)
    
    ax.set_xlim([0, 25])
    ax.set_ylim([0, 1.1])
    plt.show()
    return fig
```
